# subplot()  ---> 2 plots

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1
x = np.array([0,6,20,33,46,59])
y = np.array([11,39,42,60,81,83])
plt.subplot(1,2,1)   # use ----> subplot()   | (1,2,1) --> 1st row, 2nd column, 1st plot
plt.plot(x,y)


# plot 2
x = np.array([1,4,6,8,23,33])
y = np.array([0,22,40,46,55,67]) 
plt.subplot(1,2,1)   # use ----> subplot()     | 1,2,1) --> 1st row, 2nd column, 1st plot
plt.plot(x,y)


plt.title("x and y plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(axis="x")    # to show only x-axis
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# plot 1
x = np.array([0,6,20,33,46,59])
y = np.array([11,39,42,60,81,83])
plt.subplot(1,2,1)   # use ----> subplot()   | (1,2,1) --> 1st row, 2nd column, 1st plot
plt.plot(x,y)


# plot 2
x = np.array([1,4,6,8,23,33])
y = np.array([0,22,40,46,55,67]) 
plt.subplot(1,2,2)   # use ----> subplot()     | 1,2,2) --> 1st row, 2nd column, 2nd plot
plt.plot(x,y)


plt.title("x and y plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(axis="x")    # to show only x-axis
plt.show()
```
